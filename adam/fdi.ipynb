```python
def filt(df, label):
    df = df[(df['dif'] >= 0) & (df['dif'] <= 15)]
#     df = df[(df['dif'] >= 0)]
    df['relative_rate'] = df[label] / df['base']
    df.loc[:, 'is_2008'] = df.index == 2008
    df = df.sort_values(by='dif', ascending=True)
    return df[['relative_rate', 'dif', 'open_year' , 'announce_year', 'is_2008']]
```

```python
uk2 = filt(uk, 'total_fdi')
year_brazil2 = filt(year_brazil, 'brazil_(DC)FDI: Equity Capital (EC): Total')
china2 = filt(china, 'china_FDI: No of Contract: Total')
year_russia2 = filt(year_russia, 'russia_Foreign Direct Investments (FDI): BoP Data: Inward')
greece2 = filt(greece, 'greece_Foreign Direct Investment (FDI): Flow')
US2 = filt(US, 'US_Foreign Direct Investment (FDI)')
```

```python
dfs = [("UK 2005-2012", uk2, 'r'), 
       ("Brazil 2009-2016", year_brazil2, 'b'),
       ('China 2001-2008', china2, 'y'),
       ('Russia 2007-2014', year_russia2, 'g'),
#        ('Greece', greece2, 'c'),
       ('United States 1995-2002', US2, 'm')]
fig,ax = plt.subplots(figsize = (15, 8))
ax.set_ylim(0, 5)
for label, df, color in dfs:
    x = df.dif.values
    y = df.relative_rate
    xx = []
    yy = []
    if label == 'Greece':
        ax2=ax.twinx()
        ax2.plot(x, y, label=label, color=color, linestyle='dashed', marker='o', linewidth=1)
        continue
    for idx, row in df.iterrows():
        if row.is_2008:
            xx.append(row.dif)
            yy.append(row.relative_rate)
    if len(xx) > 0:
        ax.scatter(xx, yy, marker='^', label="2008 Financial Crisis", s=500, color = color)
        ax.plot(x, y, label=label, color=color, linestyle='dashed', marker='o', linewidth=4)
    else:
        ax.plot(x, y, label=label, color=color, linestyle='dashed', marker='o', linewidth=1)
        
ax.legend()
ax.grid()
ax.set_title("FDI(Foreign Direct Investment) relative to the FDI of announcement year for each host country", fontsize=16, color='red')
ax.set_ylabel("Ratio to the value in the announcement year", fontsize=20, color='blue')
ax.set_xlabel("relative shift to the announcement year (0 = announcement year)", fontsize=20, color='blue')
plt.vlines(8, 0, 5, linewidth=2, color='g', label="Open Date")
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
fig.show()
fig.savefig("fdi", dpi=500)
```
