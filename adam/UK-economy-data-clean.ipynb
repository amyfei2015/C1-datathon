```python
import pandas as pd
from datetime import datetime
root = "../datasets_full/UK_clean/"
```

```python
# unemployment
path = "../datasets_full/UK/Unemployment_rates_UK.csv"
df = pd.read_csv(path, skiprows=6)
```

```python
def get_time(s):
    q, y = s.split(" ")
    q = q.split("-")[-1]
    dt = datetime.strptime(q + "-" + y, "%b-%Y")
    return dt
```

```python
df['Period'] = df['Period'].map(get_time)
df['quarter'] = df['Period'].dt.quarter
df['year'] = df['Period'].dt.year
df = df.drop(columns=['Period'])
```

```python
df = df.groupby(['year', 'quarter']).mean()
df = df.rename(columns = {
    'People': 'unemployment_rate'
})
df = df[['unemployment_rate']]
```

```python
df.to_csv(root + "unemployment_rate_csv", index=True)
```

```python
# exchange rate
path = "../datasets_full/UK/USUKFXUKQ.csv"
df = pd.read_csv(path, parse_dates=['DATE'])
df['quarter'] = df['DATE'].dt.quarter
df['year'] = df['DATE'].dt.year
df = df.drop(columns=['DATE'])
df = df.rename(columns={'USUKFXUKQ': 'fx'})
df.to_csv(root+"fx.csv", index=True)
```

```python
# interest rate
path = "../datasets_full/UK/Bank Rate Bank of England Database.csv"
df = pd.read_csv(path)
df['Date Changed'] = pd.to_datetime(df['Date Changed'], format="%d %b %y")
df['quarter'] = df['Date Changed'].dt.quarter
df['year'] = df['Date Changed'].dt.year
df = df.drop(columns=['Date Changed'])
df = df.rename(columns={'Rate': 'int_rate'})
df.to_csv(root+"int_rate.csv", index=True)
df.head()
```
